## Question 2 (10 Points)
Write python code to extract these features you discussed above. You can collect a few sample text data for the feature extraction.

```python
# You code here (Please add comments in the code):
# Import Required libraries
import nltk
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk import pos_tag
from nltk.sentiment import SentimentIntensityAnalyzer
import pandas as pd


#  Sample data of the product reviews
reviews = [
    "I absolutely love this product! It works amazingly well.",
    "The product is okay, but not what I had hoped for.",
    "Not happy with the quality, I expected something better.",
    "This is a terrible product, I will never buy it again.",
    "Fantastic quality! Highly recommend to everyone."
]

# 1. Bag of Words (BoW) feature extraction
vectorizer = CountVectorizer()
# Reviews fitting and transformation
R_bow = vectorizer.fit_transform(reviews)
print("1. Bag of Words Features:\n", R_bow.toarray())

# 2. Sentiment Lexicon (Using VADER in performing the analysis)
nltk.download('vader_lexicon')
senta = SentimentIntensityAnalyzer()
sentiment_scores = [senta.polarity_scores(review) for review in reviews]
print("\n2. Sentiment Lexicon Features (VADER scores):")
for review, score in zip(reviews, sentiment_scores):
    print(f"Review: {review}\nSentiment Scores: {score}\n")

# 3. Part of Speech (POS) Tags
#Downloading Tokenizer- a model used for splitting text into words/sentences
nltk.download('punkt')
#Downloading model for tagging parts of the speech
nltk.download('averaged_perceptron_tagger')
print(" 3. POS Tags for Each text given in the Review")
for review in reviews:
    tokens = word_tokenize(review)
    pos_tags = pos_tag(tokens)
    print(f"Review: {review}\nPOS Tags: {pos_tags} \n")
# 4. TF-IDF
tfidf_vectorizer = TfidfVectorizer(stop_words='english', ngram_range=(2, 3))
tfidf_matrix = tfidf_vectorizer.fit_transform(reviews)
print("\n 4. TF-IDF Matrix using (Bigrams & Trigrams):\n",)
print(tfidf_vectorizer.get_feature_names_out())
print(tfidf_matrix.toarray())

# 5. N-grams (Bigrams & Trigrams)
# CountVectorizer to generate bigrams and trigrams (n-gram range from 2 to 3)
vectorizer = CountVectorizer(ngram_range=(2, 3), stop_words='english')
ngrams = vectorizer.fit_transform(reviews)
print("Bigrams and Trigrams (2-grams & 3-grams):")
print(vectorizer.get_feature_names_out())
print(ngrams.toarray())
```

## Question 3 (10 points):
Use any of the feature selection methods mentioned in this paper "Deng, X., Li, Y., Weng, J., & Zhang, J. (2019). Feature selection for text classification: A review. Multimedia Tools & Applications, 78(3)."

Select the most important features you extracted above, rank the features based on their importance in the descending order.

```python
# You code here (Please add comments in the code):

# Import necessary libraries
from sklearn.feature_selection import chi2
from sklearn.feature_extraction.text import CountVectorizer
import numpy as np

# Labels for the reviews (1 for positive, 0 for negative)
labels = [1, 0, 0, 0, 1]  # Positive = 1, Negative = 0

# 1. Bigrams and Trigrams Feature Extraction
vectorizer = CountVectorizer(ngram_range=(2, 3), stop_words='english')
X = vectorizer.fit_transform(reviews)
feature_names = vectorizer.get_feature_names_out()

# 2. Apply Chi-Square for Feature Selection
chi2_scores, p_values = chi2(X, labels)

# 3. Ranking the features by Chi-Square score
# Sorting features by the highest chi2 score (descending order)
sorted_idx = np.argsort(chi2_scores)[::-1]
sorted_feature_names = feature_names[sorted_idx]
sorted_chi2_scores = chi2_scores[sorted_idx]

# Showing top features according to its essense
print("Top Features Ranked by Importance (Chi-Square Score):")
print(" \n Top Features.............Scores\n")
for feature, score in zip(sorted_feature_names, sorted_chi2_scores):

     print(f"{feature}: {score:.2f}")



```

## Question 4 (10 points):
Write python code to rank the text based on text similarity. Based on the text data you used for question 2, design a query to match the most relevant docments. Please use the BERT model to represent both your query and the text data, then calculate the cosine similarity between the query and each text in your data. Rank the similary with descending order.

```python
# You code here (Please add comments in the code):

#importing the required libraries
from transformers import BertTokenizer, BertModel
import torch
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

# Below is the query that I want to search for within the reviews
query = "Great quality and highly recommend"


# 1. Loading pre-trained BERT model and tokenizer
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
model = BertModel.from_pretrained('bert-base-uncased')

def get_bert_embedding(text):
    """
    Generate BERT embeddings for a given text.
    """
    inputs = tokenizer(text, return_tensors='pt', padding=True, truncation=True, max_length=512)
    with torch.no_grad(): #Disabling the calculation of the gradient.
        outputs = model(**inputs)
# Getting the average embedding
    return outputs.last_hidden_state.mean(dim=1)

# 2. Getting BERT embeddings for all reviews and the query
query_embedding = get_bert_embedding(query)
review_embeddings = [get_bert_embedding(review) for review in reviews]

# 3. Cosine similarity between query embedding and each review
similars = [cosine_similarity(query_embedding, review_emb)[0][0] for review_emb in review_embeddings]

# 4. Ranking reviews by similarity in descending order
sorted_idx = np.argsort(similars)[::-1]  # Indices of reviews sorted by similarity

# 5. Displaying ranked results
print("Ranked Reviews by Similarity to Query Designed:")
for idx in sorted_idx:
    print(f"Review: {reviews[idx]}, Similarity: {similars[idx]:.4f}")






```
